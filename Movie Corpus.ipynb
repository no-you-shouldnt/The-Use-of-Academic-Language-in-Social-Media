## ABRIR TEXTOS

```python
with open('movies-text/movies_text_CUT.txt') as f:
    contents=f.read()
```

## PREPROCESO

```python
#eliminar () y []
def a(test_str):
    ret = ''
    skip1c = 0
    skip2c = 0
    for i in test_str:
        if i == '[':
            skip1c += 1
        elif i == '(':
            skip2c += 1
        elif i == ']' and skip1c > 0:
            skip1c -= 1
        elif i == ')'and skip2c > 0:
            skip2c -= 1
        elif skip1c == 0 and skip2c == 0:
            ret += i
    return ret
```

```python
contents = a(contents)
```

```python
contents= contents.split('.')
```

```python
#importando librerias
import spacy
import nltk
import string
from nltk import sent_tokenize
import numpy as np
```

```python
#dividir en oraciones, importar la funcion
from nltk import sent_tokenize
```

```python
#dividir en oraciones
sentences = []
for sentence in contents:
    sent_token = sent_tokenize(sentence)
    sentences+=[sent_token]
```

```python
sentences[:15]
```

```python
##eliminar espacios en blanco
sentences = [palabra for palabra in sentences if len(palabra)>0]
```

```python
## tokens y types

tokens = []
for word in sentences:
    tokens+=sentence.split(' ')

tokens = [word.lower() for word in tokens]
```

```python
#aplanar la lista de strings
flat_sentences = []
for item in sentences:
        for word in item:
            flat_sentences+=[word]
```

```python
#remover puntuación
import re

def remove_punctuation(s):
    return re.sub(r'[^\w\s]','',s)
```

```python
flat_sentences_nopunc =[]

for sent in flat_sentences:
    flat_sentences_nopunc+=[remove_punctuation(sent)]
```

## NOMINALIZACIONES

```python
## identificamos el lema y el pos

import spacy
nlp = spacy.load("en_core_web_sm")
oraciones_nlp = []

for sentence_text in flat_sentences_nopunc:
    if len(sentence_text)>1:
        sent = []
        for sentence in sentence_text:
            doc = nlp(sentence)
            sent += [(token.lemma_.lower(),token.pos_) for token in doc]
        oraciones_nlp += [sent]
```

```python
## reglas de nominalization
## puse las palabras lematizadas, en ese caso, solo es necesario poner las palabras en singular

no_nom = ['thing', 'things', 'somethings', 'something', 'anything', 'everything', 'nothing','original', 'special', 'normal', 'version', 'tutorial', 'moment', 'criminal', 'tradition', 'morning', 'question', 'element', 'quality']
terminacion = ['ment', 'tion', 'sion', 'ibility', 'ity', 'ness', 'al', 'ing'] 
```

```python
#oraciones con POSIBLES nominalizaciones

def nominalization(oraciones):
    nom = []
    for word in oraciones:
        if word not in no_nom:
            if word=='NOUN':
                for END in terminacion:
                    if word.endswith(END):
                        nom+=[word]
    return nom
```

```python
noms = []
noms = nominalization(flat_sentences_nopunc)
```

```python
noms
```

```python
##porcentaje nominalizaciones

len(noms)/len(sentences)*100
```

```python
#dividir por salto de linea
AWL= AWL.split('\n')
```

```python
##palabras uncias

for palabra in AWL:
    if palabra not in AWL:
        AWL += [palabra]
```

```python
##eliminar espacios en blanco
AWL = [palabra for palabra in AWL if len(palabra)>0]
AWL[:10]
```

```python
def academic(tokens):
    aca_list=[]
    for word in tokens:
        if word in AWL:
            aca_list+=[word]
    return aca_list
```
